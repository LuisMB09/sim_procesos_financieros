```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
df.info()
```

```python
# convert object columns to category for better memory usage
df['gender'] = df['gender'].astype('category')
df['smoking_history'] = df['smoking_history'].astype('category')

# convert binary indicators to boolean type
df['hypertension'] = df['hypertension'].astype(bool)
df['heart_disease'] = df['heart_disease'].astype(bool)
df['diabetes'] = df['diabetes'].astype(bool)

# float
df['age'] = df['age'].astype(np.float32)
df['bmi'] = df['bmi'].astype(np.float32)
```

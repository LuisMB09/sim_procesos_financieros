```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
k = 50
p = 5
```

```python
def long_call_payoff(st, k, premium):
    return np.maximum(st - k, 0) - premium

def long_put_payoff(st, k, premium):
    return np.maximum(k - st, 0) - premium
```

```python
st = np.linspace(0, 200, 10_000)
```

```python
call = long_call_payoff(st, k, p)
put = long_put_payoff(st, k, p)

straddle = call + put
```

```python
plt.figure(figsize=(10, 6))
plt.plot(st, call, label='Call Option Payoff', color='blue')
plt.plot(st, put, label='Put Option Payoff', color='red')
plt.plot(st, straddle, label='Straddle Payoff', color='green')
plt.title('Long Straddle Payoff Diagram')
plt.xlabel('Stock Price at Expiration ($)')
plt.ylabel('Payoff ($)')
plt.legend()
plt.grid()
plt.show()
```
